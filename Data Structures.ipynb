```python
##Recursive maximum contiguous subsequence sum algorithm.
##Finds maximum sum in subarray spanning a[left..right].
##Does not attempt to maintain actual best sequence.
def maxSumRec( a,left,right ):
    print('left',left,'right',right)
    if( left == right ):
        if( a[ left ] > 0 ):
            return a[ left ]
        else:
            return 0
    center = ( left + right ) / 2
    print ('Before maxLeftSum Interation:center',center,'left',left,'right',right)
    maxLeftSum = maxSumRec( a, left, center ) 
    print ('After maxLeftSum Interation:center',center,'left',left,'right',right)
    print ('maxLeftSum',maxLeftSum)
    print ('Before maxRightSum Interation:center',center,'left',left,'right',right)
    maxRightSum = maxSumRec( a, center + 1, right )
    print ('After maxRightSum Interation:right-center',center+1,'left',left,'right',right)
    print ('maxRightSum',maxRightSum)
    maxLeftBorderSum =leftBorderSum = 0
    i=center
    print('center',center,'right',right,'left',left)
    while i>=left:
        leftBorderSum =leftBorderSum+ a[ i ]
        if( leftBorderSum > maxLeftBorderSum ):
            maxLeftBorderSum = leftBorderSum
        i=i-1
    maxRightBorderSum = 0
    rightBorderSum = 0
    i=center + 1
    print('center+1',center+1,'right',right,'left',left)
    for i in range(center + 1,right)[::1]:
        rightBorderSum = rightBorderSum+a[ i ]
        if( rightBorderSum > maxRightBorderSum ):
            maxRightBorderSum = rightBorderSum
    print('maxLeftSum',maxLeftSum, 'maxRightSum',maxRightSum, 'maxLeftBorderSum: maxRightBorderSum',maxLeftBorderSum , maxRightBorderSum)
    return max( maxLeftSum, maxRightSum, maxLeftBorderSum + maxRightBorderSum )
##Driver for divide-and-conquer maximum contiguous
##subsequence sum algorithm.
def maxSubSum3(a):
    return maxSumRec( a, 0, len(a)- 1 )
    

def main():
    a=[4,-3,5,-2,-1,2,6,-2]
    k=maxSubSum3(a)

if __name__ == '__main__':
    main()
```

```python
# quick sort
def partition(myList, start, end):
    pivot = myList[start]
    left = start+1
    # Start outside the area to be partitioned
    right = end
    done = False
    while not done:
        while left <= right and myList[left] <= pivot:
            left = left + 1
        while myList[right] >= pivot and right >=left:
            right = right -1
        if right < left:
            done= True
        else:
            # swap places
            temp=myList[left]
            myList[left]=myList[right]
            myList[right]=temp
    # swap start with myList[right]
    temp=myList[start]
    myList[start]=myList[right]
    myList[right]=temp
    return right


def quicksort(myList, start, end):
    if start < end:
        # partition the list
        split = partition(myList, start, end)
        # sort both halves
        quicksort(myList, start, split-1)
        quicksort(myList, split+1, end)
    return myList



def main():
    myList = map(int,raw_input().split())
    sortedList = quicksort(myList,0,len(myList)-1)
    print(sortedList)

if __name__ == '__main__':
    main()
```
